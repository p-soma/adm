```python
import pandas as pd
data = pd.read_csv("CapstoneData.csv",error_bad_lines=False)
data.head()
```

```python
from sklearn.model_selection import train_test_split
data = data.reset_index()
Y = data["FraudFlag"]
X= data.drop(['OURACCOUNT',
               'LOCALTRANTIME',
               'LOCALTRANDATE',
               'TRACENUMBER',
               'REFERENCENUMBER',
               'OURTRANSMISSIONTIME',
               'OURTRANSMISSIONDATE',
              'CARDACCEPTORCITY',
              'CARDACCEPTORCOUNTRY',
              'CARDACCEPTORIDCODE',
              'CARDACCEPTORNAME',
              'CARDACCEPTORSTATE',
              'CARDACCEPTORSTREET',
              'RESPONSECODEIN',
              'TIME',
              'OURTRANCODE',
              'AUTHIDRESPONSE'
                   ], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
X_train.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

rng = np.random.RandomState(42)

clf = IsolationForest(random_state=rng, )

clf.fit(X_train, y_train)
y_pred_train = clf.predict(X_train)
y_pred_test = clf.predict(X_test)

```

```python
from sklearn.metrics import confusion_matrix
test=[]
for val in y_pred_test:
    if val == 1:
        test.append(0)
    else:
        test.append(1)

confusion_matrix(y_test,test)
```
